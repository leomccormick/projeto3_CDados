### LEITURA DE APENAS ALGUMAS COLUNAS <br>
Para saber o significado das linhas e colunas, veja o arquivo Excel `dicionario_PNADC_microdados_trimestral.xlsx`

```python
# Colunas selecionadas, POR EXEMPLO:
colunas = ['V403312', # Salario
            'UF',     # Estado
           'V2010',   # Cor ou raça
           'V3001',   # Sabe ler ou escrever?
           'VD3004',  # Nível de instrução
           'V4029',   # Tem carteira assinada?
           'V1023',   # Tipo de área da residência
           'V2009',   # Idade  
           'V4012',   # Tipo de trabalho
           'V4020']   # Local do trabalho

# Leitura da base de dados filtrando apenas as colunas contidas na lista `colunas`
pnad23_1_small = pd.read_csv(arquivo_local, usecols=colunas)
```

```python
pnad23_1_small.rename(columns={'V403312':'Salario',
                                'UF':'Estado',
                                'V2010':'CorRaca',
                                'V3001':'Analfabeto',
                                'VD3004':'Instrucao',
                                'V4029':'Carteira',
                                'V1023':'AreaRes',
                                'V2009':'Idade',
                                'V4012': 'TipoTrabalho',
                                'V4020': 'LocalTrabalho'}, inplace=True)
pnad23_1_small
```

```python
pnad23_1_small.to_csv("output.csv", index=False)
```
